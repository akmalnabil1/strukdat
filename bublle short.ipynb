Pengertian Bubble Sort Bubble Sort adalah algoritma pengurutan sederhana yang bekerja dengan membandingkan dua elemen bersebelahan dan menukarnya jika urutannya salah. Proses ini dilakukan berulang kali hingga tidak ada lagi pertukaran.

Langkah-Langkah Bubble Sort a.Bandingkan dua elemen yang bersebelahan. b.Tukar jika urutannya tidak sesuai. c.Lanjutkan hingga akhir daftar. d.Ulangi proses sampai data terurut.



Contoh Data: 5, 3, 8, 4, 2 Hasil Akhir (Ascending): 2, 3, 4, 5, 8




Kelebihan dan Kekurangan Kelebihan: a.Mudah dipahami dan diimplementasikan. b.Cocok untuk data kecil. c.Stabil dan tidak membutuhkan memori tambahan.


Kekurangan: Tidak efisien untuk data besar. Banyak melakukan pertukaran.


Kesimpulan Bubble Sort adalah algoritma pengurutan sederhana dan mudah dipelajari. Cocok digunakan untuk pembelajaran dasar dan dataset kecil. Memiliki kompleksitas O(n²), sehingga kurang efisien untuk data besar. Dapat dioptimalkan untuk meningkatkan performa pada data yang hampir terurut.

```python

lst = [8,5,3,1]

def bubble_sort_ascending(arr):
    n = len(arr)
    for i in range(n - 1):
        print(lst)
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

print(bubble_sort_ascending(lst))

lst = [4,7,1,9,3,10,6]
print(lst)

for i in range(0, len(lst)-1):
    print(i)
    min = i
    berubah = False  # flag perubahan

    for j in range(i+1, len(lst)):
        if lst[j] < lst[min]:
            min = j

    # cek apakah perlu swap
    if min != i:
        lst[i], lst[min] = lst[min], lst[i]
        berubah = True  # terjadi perubahan

    # tampilkan hanya jika berubah
    if berubah:
        print(lst)

def insertion_sort_ascending(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
    return arr

data = [8, 3, 5, 2]
print("Ascending:", insertion_sort_ascending(data.copy()))

#-------------------------------------------------------

def insertion_sort_descending(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] < key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
    return arr

data = [8, 3, 5, 2]
print("Descending:", insertion_sort_descending(data.copy()))

#---------------------------------------------------------------

nama = ["Zahra", "Andi", "Budi", "Citra", "Dina"]

def insertion_sort_nama(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j].lower() > key.lower():
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
    return arr

print("Urutan Alfabet:", insertion_sort_nama(nama))

#-------------------------------------------------------

def insertion_sort_optimized(arr):
    """
    Insertion Sort dengan optimalisasi.
    Menggunakan pergeseran elemen dan berhenti lebih awal
    jika posisi elemen sudah benar.
    """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # Jika elemen sudah pada posisi yang tepat, lewati
        if arr[j] <= key:
            continue

        # Geser elemen yang lebih besar ke kanan
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        # Sisipkan elemen pada posisi yang benar
        arr[j + 1] = key

    return arr


# Contoh penggunaan
data = [64, 25, 12, 22, 11]
print("Insertion Sort Optimized:", insertion_sort_optimized(data.copy()))
```
