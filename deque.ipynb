Deque, atau double-ended queue, adalah tipe struktur data yang mirip dengan queue dan stack, namun dengan kemampuan untuk menambahkan dan menghapus elemen di kedua ujungnya. Dalam deque, kita dapat menambahkan elemen baru baik di depan maupun di belakang, serta menghapus elemen dari depan dan belakang pula.

```python
def createDeque():
    d=[]
    return (d)

def addFront(d,data):
    d.insert(0,data)
    return(d)

def addRear(d,data):
    d.append(data)
    return(d)

def removeRear(d):
    data=d.pop()
    return(data)

def removeFront(d):
    data=d.pop(0)
    return(data)

def isEmpty(d):
    return (d==[])

def size(d):
    return (len(d))

# contoh sederhana penggunaan fungsi fungsi yang terdapat pada deque

d = createDeque()

addRear(d,10)
addRear(d,20)
addRear(d,30)
addRear(d,40)
addRear(d,50)

print("Deque awal:", d)

removeFront(d)
removeFront(d)
removeRear(d)

print("Deque akhir:", d)
```
