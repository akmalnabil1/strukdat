Python memiliki tipe data dinamis yaitu List, yang ukurannya bisa berubah saat program berjalan dan dapat ditambah datanya kapan saja menggunakan method bawaan. Namun, tidak semua bahasa pemrograman menyediakan tipe data dinamis seperti ini. Karena itu, digunakan struktur data Linked List, yaitu struktur data dinamis yang dibuat sendiri oleh programmer. Hal utama dalam struktur data Linked List:

Node → elemen penyimpan data dan alamat node berikutnya.

Penambahan Data → proses memasukkan node baru ke linked list.

Traversal → proses menelusuri atau membaca seluruh isi linked list.

Ordered List → linked list yang datanya tersusun secara terurut. NODE Pada Linked List, data disimpan di alamat memori yang berbeda-beda. Agar data saling terhubung, setiap data membutuhkan informasi tambahan berupa link/alamat node berikutnya. Data pertama disebut head dan harus diketahui secara eksplisit melalui external reference. Data terakhir memiliki penanda bahwa tidak ada node berikutnya lagi.

```python
# pembuatan class node
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None
    def getData(self):
        return self.data
    def getNext(self):
        return self.next
    def setData(self, newdata):
        self.data = newdata
    def setNext(self, new_next):
        self.next = new_next
a=Node(93)
b=Node(20)
print(a.getNext())
print(b.getNext())
a.setNext(b)
print(a.getNext())
```

Linked List adalah kumpulan node yang saling terhubung. Untuk mengakses seluruh node dalam linked list, harus diketahui lokasi node pertama (head). Karena itu diperlukan pointer tambahan untuk menunjukkan keberadaan node pertama tersebut. Class LinkedList memiliki pointer head yang menunjuk ke node pertama dalam linked list. Pada contoh juga terdapat class Node yang memiliki data dan pointer ke node berikutnya (next). Method penting pada class LinkedList

```python
# kode linnked lsit class
class LinkedList:
    def __init__(self):
        self.head = None
    def isEmpty(self):
        return self.head==None
mylist=LinkedList()
print(mylist.head)
mylist.isEmpty()
print(mylist.head)
mylist.isEmpty()
```
